| **Operator** | **Purpose** |
|:----------|:---------|
| `==` | x is  equal to y |
| `!=` | x is not equal to y |
| `x > y` | x is greater than y |
| `x < y` | x is less than y |
| `x >= y` | x is greater than or equal to y | 
| `x <= y` | x is less than or equal to y | 

| Operator | Technical name | Math symbol | Python construct | 
|:--------:|:--------------:|:------:|:------:|
| And      | Conjunction    | $\land$| `and`  |
| Or       | Disjunction    | $\lor$ | `or`   |
| Not      | Negation       | $\lnot$| `not`  |



The order is as follows: `not` > `and` > `or`, again if in doubt *use brackets!*

```python
#examples:
print(not False and True)
print(True or False and True)
```

There are 2 operators equivalent to== and != namely is and is not, respectively.

* Chained and nested conditionals

```python
x: int = 4
y: int = 4

if  x < y:
    print('x is less than y')
elif x > y:
    print('x is greater than y')
else:
    print('x and y are equal')    
```

```python
#aviod nested conditionals when possible
x: int = 2
y: int = 2

if x == y:
    print('x and y are equal')
else:
    if x < y:
        print('x is less than y')
    else:
        print('x is greater than y')
```

```python
#example:
vowels: str = 'aeiou'
word: str = 'supercalifragilisticexpialidocious'

vowel_count: int = 0  

for letter in word:        #it iterates over all the characters in the word. 
    if letter in vowels:     
        vowel_count += 1     

print(vowel_count)
```
